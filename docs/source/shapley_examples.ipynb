```python

import pyciemss
import pyciemss.visuals.plots as plots


import xgboost
import xgboost as xgb
import shap
import pandas as pd
import numpy as np
import json
```

```python
# train XGBoost model
X, y = shap.datasets.adult(n_points=20)
model = xgboost.XGBClassifier().fit(X, y)
explainer = shap.Explainer(model, X)
shap_values = explainer(X)
```

```python
new_schema = plots.shapley_decision_plot(shap_values)

with open('decision_schema.json', 'w') as f:
    json.dump(new_schema, f)

plots.ipy_display(new_schema, dpi=150)
```
