```python
import pyciemss
```

```python
start_time = 0.0
end_time = 40.0
logging_step_size = 1.0
num_samples = 1

model = "SIR_with_observable_namespace_collision.json"

result_preoptimize = pyciemss.sample(
    model,
    end_time,
    logging_step_size,
    num_samples,
    start_time = start_time,
    solver_method = "dopri5"
)
```

```python
model = "SIR_with_observable.json" # has observable Rnot instead of R0

result_preoptimize = pyciemss.sample(
    model,
    end_time,
    logging_step_size,
    num_samples,
    start_time = start_time,
    solver_method = "dopri5"
)
display(result_preoptimize["data"])
```
